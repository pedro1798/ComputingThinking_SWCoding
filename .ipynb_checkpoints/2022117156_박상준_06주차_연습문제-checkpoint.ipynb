# a < b라는 조건이 항상 지켜진다고 하자.
# 여러 수의 최소공배수 구하여라

# a,b: 최소공배수를 구하고자 하는 두 수
# gcd(a,b): a와b의 최대공약수
# (최소공배수 * 최대공약수 = a * b)
# 최소공배수 = a * b / 최대공약수(gcd)
# 여러 수의 최소공배수 : 두 수의 최소공배수로 나온 값 다음 수와 다시 최소공배수 구하면 됨
def gcd(a, b): #greatest common divisor 최대공약수 # a < b
    c = 0 # 12 15
    while a != 0:
        c = b % a # c = 3  c = 0
        b = a # b = 12 b = 3
        a = c # a = 3 a = 0 루프 탈출
    return b # 최대공약수 리턴

def lcm(a, b):
    return a * b // gcd(a, b)

a = int(input("범위의 시작 정수 : "))
b = int(input("범위의 마지막 정수 : "))
c = [] # 2 3 4
for i in range(a, b+1):
    c.append(i)

d = c[0]
for i in range(1, len(c)):
    d = lcm(d, c[i]) # 0과 1의 최소공배수 구해 lcm에 저장

print(f"{a}에서 {b}까지의 정수들의 최소 공배수는 : {d}")
</code></pre>

# m과 n이라는 이름의 매개변수를 두개 입력받아서
# 이 두 값 중에서 큰 값을 반환하는 max2(m,n) 함수와
# 작은 값을 반환하는 min2(m,n) 함수를 구현하시오.
# 100과 200을 인자로 넣어서 호출하여 그 결과를 확인하시오.

def max2(m, n):
    if m > n:
        print(f"{m}과 {n}중 큰 수는 : {m}")
    elif m < n:
        print(f"{m}과 {n}중 큰 수는 : {n}")
    else:
        print(f"{m}과 {n}은 같다.")
def min2(m, n):
    if m > n:
        print(f"{m}과 {n}중 작은 수는 : {n}")
    elif m < n:
        print(f"{m}과 {n}중 작은 수는 : {m}")
    else:
        print(f"{m}과 {n}은 같다.")
max2(100, 200)
min2(100, 200)
