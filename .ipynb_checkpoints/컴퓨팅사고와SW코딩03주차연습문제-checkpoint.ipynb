```python
#1, 세 개의 정수 받아서 작은 순대로 정렬 if-else 문으로
# 세 정수는 모두 다르다고 가정
"""Another strategy ', '.join(c for n, c in sorted(zip([a, b, c], 'abc'), reverse=True)).
This pairs numbers with letter using zip, sorts them largest to smallest,
and then takes just the letters. –
Steven Rumbalski
 4 mins ago """
a, b, c = map(int, input("3개의 정수를 입력하시오 : ").split())

if c < a > b:
    print((c, b, a) if c < b else (b, c, a))
elif a < b > c:
    print((a, c, b) if a < c else (c, a, b))
elif a < c > b:
    print((a, b, c) if a < b else (b, a, c))
else:
    print("wrong code!!!")
    
"""
note : 
https://stackoverflow.com/questions/73735398/how-to-compare-three-numbers-with-if-else#comment130204333_73735398

"""
```

```python
# 2,사용자로부터 x, y 좌표를 가진 한 점을 입력받아서
# 이 점이 1, 2, 3, 4 분면의 어디에 속하는지를 알려주는 프로그램을 작성하시오.

x, y = map(int, input("점의 좌표 x, y를 입력하시오 : ").split())

if x > 0 and y > 0:
    print("1사분면에 있음.")
elif x < 0 and y > 0:
    print("2사분면에 있음.")
elif x < 0 and y < 0:
    print("3사분면에 있음.")
elif x > 0 and y < 0:
    print("4사분면에 있음.")
else:
    print("원점(0, 0)에 있음.")
```

```python
#6
"""
암스트롱 수란 세 자리의 정수 중에서 각 자리의 수를 세제곱한 수의 합과 자신이 같은 수를 말한다. 
구체적인 예를 들면 153은 1 + 125 + 27 의 합으로 구성될 수 있는데 이러한 수가 암스트롱 수이다. 
세 자리 정수들 중에서 모든 암스트롱 수를 구하여 다음과 같이 출력하여라.
"""
print("세 자리의 암스트롱 수 : ", end = "")
for i in range(100, 1000):
    #각 자리수 분리하기, 380
    a = i // 100
    b = (i - a * 100) // 10
    c = (i - a * 100 - b * 10) // 1
    amstrong = a*a*a + b*b*b + c*c*c
    if amstrong == i:
        print(i, end=" ")
```

```python
#7
"""
다음과 같이 사용자로부터 1에서 9사이의 숫자를 입력받아
입력받은 숫자의 절에 해당하는 구구단을 출력하는 프로그램을 for문과 while 문을 사용하여 작성하여라.
만일 1에서 9 이외의 숫자가 입력되면 ‘1에서 9까지의 수를 다시 입력하세요’라는 안내문을 출력하여라.
"""

n = int(input("1에서 9까지의 수를 입력하세요 : "))
if n < 1 or n > 9 :
    while(n < 1 or n > 9):
        n = int(input("1에서 9까지의 수를 다시 입력하세요 : "))

for i in range(1,10):
    print(f"{n} * {i} = ", n * i)
```

```python
#8
"""
맛나 식당의 메뉴 주문 프로그램을 개발하고자 한다. 
이를 위해 사용자에게 다음과 같은 메뉴를 보여주고 이중에서 메뉴에 대응되는 하나의 알파벳을 선택하도록 하자. 
이때 메뉴에 없는 알파벳이 입력되면 ‘메뉴를 다시 입력하세요:  ’가 출력되도록 한 후 다시 입력을 받도록 하자.
"""
from operator import eq

print("맛나 식당에 오신 것을 환영합니다. 메뉴는 다음과 같습니다.\n"
      "-삼겹구이 (입력 s)\n"
      "-오삼불고기 (입력 o)\n"
      "-된장찌개 (입력 d)")
a = input("메뉴를 선택하세요 ( 알파벳 s, o, d 입력 ) : ")
while(1):
    if eq(a, "s"):
        print("삼겹구이를 선택하였습니다.")
        break
    elif eq(a, "o"):
        print("오삼불고기를 선택하였습니다.")
        break
    elif eq(a, "d"):
        print("된장찌개를 선택하였습니다.")
        break
    else:
        a = input("메뉴를 다시 입력하세요 : ")

# == 로는 문자열 비교가 되지 않는것 같아 operator 모듈의 eq() 함수를 이용하였습니다.
```

```python
#9
"""
이중 for 문을 사용하여 숫자를 입력 받아 다음과 같은 삼각형을 출력하는 프로그램을 작성하시오. 
이때 아래 결과와 같이 숫자 10이 입력되면 높이가 10이고 제일 마지막 줄의 별이 10개가 삼각형 형태로 나타나도록 하시오.
힌트: 첫번째줄은 공백 9개, 별1개 / 마지막줄은 공백이 없다. 
"""
n = int(input("숫자를 입력하세요 : "))
for i in range(1, n):
    for j in range(0, n - i):
        print(" " ,end="") # n - i 만큼 공백찍기 -> n - i 반큼 for문 반복
    print("*" * i)
```

```python
#10
"""
사용자로부터 숫자 1보다 크고 10보다 작은 값 n을 입력으로 받아서 
다음과 같이 뱀의 몸통처럼 증가하는 이차원 배열을 출력하여라. 
(이 배열은 마치 뱀의 몸통처럼 ‘ㄹ’과 같은 패턴을 그리며 수가 1씩 증가하는 형태의 배열이다.)
"""

"""
1   2  3  4   5           1   2   3   4
10  9  8  7   6           8   7   6   5
11 12 13 14 15
"""
n = int(input("숫자를 입력하세요 : "))
cols = n
rows = n
arr = [[0 for j in range(cols)] for i in range(rows)]
#cols 는 열 ( 세로 ), rows 는 행 ( 가로 )
for i in range(n):
    for j in range(1, n+1):
        arr[i][j-1] = j + i*n
for i in range(1, n+1):
    if i % 2 == 0:
        arr[i-1].sort(reverse=True)
    print(*arr[i-1], sep=" ")

"""
2차원 배열로 풀면 좋을 것 같아 우선 nxn 크기의 2차원 배열을 만들어 숫자를 1부터 n까지 할당하고 
짝수번째 행마다 배열을 sort() 함수를 이용해 내림차순으로 정렬했습니다.
정렬 후 출력을 해 보았는데 그냥 배열을 출력하면 []가 추가되어 문제의 예시와 달라 찾아보니
*args가 argument만 받게 해주는 가변 인자를 위한 변수라는 정보가 있어
코드에 적용하였더니 정상적으로 작동하였습니다.
"""
```
