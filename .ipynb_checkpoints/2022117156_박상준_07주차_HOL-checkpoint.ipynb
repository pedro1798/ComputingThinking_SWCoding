<pre><code>
import turtle as t
import random as r
t.shape("turtle")
t.shapesize(1, 1)
t.setup(400,400)
while(1):
    a = r.randrange(0, 200)
    b = r.randrange(0, 200)
    t.penup()
    t.goto(a, b)
    t.pendown()
    t.stamp()
t.done()
</code></pre>

#### 2.

<pre><code>
import turtle as t
import random as r
t.shape("turtle")
t.shapesize(3, 3)
t.setup(400,400)
c = ['red', 'blue', 'green', 'orange']

while(1):
    a = r.randrange(-200, 200)
    b = r.randrange(-200, 200)
    t.penup()
    t.goto(a, b)
    t.pendown()
    d = r.randrange(0, 4)
    t.color(c[d])
    t.stamp()
t.done()
</code></pre>

from tkinter import *

def entry_to_label():
    str = input_entry.get()
    label.config(text=str)

window = Tk()

input_entry = Entry(window, width=50)
input_entry.pack()

button = Button(window, text = '처리', command=entry_to_label)
button.pack()

label = Label(window, text = ' ')
label.pack()

window.mainloop()
